```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

2. Create a program codes that demonstrate how to use dynamic programming to solve the same problem

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [30, 50, 100]
wt = [5, 10, 15]
w = 25
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
def bu_dp(w, wt, val, n):
    table = [[0 for x in range(w+1)] for x in range(n+1)]

    for i in range(n+1):
        for w in range(w+1):
            if i==0 or w == 0:
                table[i][w] = 0

            elif wt[i-1] <= w:
                table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]], table[i-1][w])

            else:
                table[i][w] = table[i-1][w]


    return table[n][w]

val = [30, 50, 100]
wt = [5, 10, 15]
w = 50
n = len(val)

print(bu_dp(w, wt, val, n))
```

4. Create a sample program codes that simulate tops-down dynamic programming

```python
val = [30, 50, 100]
wt = [5, 10, 15]
w = 50
n = len(val)

calc = [[-1 for i in range(w+1)] for j in range(n+1)]

def td_dp(wt, val, w, n):
    if n==0 or w==0:
        return 0

    if calc[n][w] != -1:
        return calc[n][w]

    if wt[n-1] <= w:
        calc[n][w] = max(
            val[n-1] + td_dp(wt, val, w-wt[n-1], n-1),
            td_dp(wt, val, w, n-1)
        )
        return calc[n][w]
    elif wt[n-1] > w:
        calc[n][w] = td_dp(wt, val, w, n-1)
        return calc[n][w]

print(td_dp(wt, val, w, n))
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]

  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

**Code Analysis**

My analysis for each code:
1. Recursion
Recursion solves the problem by exploring all possible items. The function rec_knapsack first checks if  n or w is 0. if it is, return 0. if an item's weight is greater than w, it will call itself again without the item. if the item is not greater, it will include the item and as stated by the code "#return the maximum of the two cases: (1) include the nth item, (2) don't include the nth item". Recursion is very simple but it is inefficient for large datasets because of repeated calculations.
2. Dynamic Programming
This approach does not use recursion. It builds a table from the smallest problem up to the final solution (like tabulation). It first creates an initialized table with 0s, and populates the table row by row. If the item is too heavy than the current capacity, it will copy the value from the cell above. Dynamic calculates only once, writes it in the table, and looks it up when it is needed for the final calculation. However, it avoids recursive calls and stack limits.
3. Memoization
This technique optimizes the recursive solution by storing the results and returning it when the same input occurs again. I think it is a combination of both recursion and dynamic programming. it keeps the simplicity of recursion while adding memory, allowing to skip repeated calculations. While it is solving, if the recursion loops back again, it will re-do from scratch. Memoization calculates only those that are needed, it skips repeated calculations.

Task 1: Modify the three techniques to include additional criterion in the knapsack problems

```python
val = [60, 100, 120]  # Value
wt  = [10, 20, 30]    # Weight
vol = [15, 20, 30]    # Volume
W   = 50              # Max Weight
V   = 40              # Max Volume
n   = len(val)
```

```python
#Recursion
def recursion(w, v, wt, vol, val, n):

    if n == 0 or w == 0 or v == 0:
        return 0


    if wt[n-1] > w or vol[n-1] > v:
        return recursion(w, v, wt, vol, val, n-1)


    else:
        return max(

            val[n-1] + recursion(w - wt[n-1], v - vol[n-1], wt, vol, val, n-1),

            recursion(w, v, wt, vol, val, n-1)
        )
```

```python
#Dynamic
def dynamicprogramming(W, V, wt, vol, val, n):

    dp = [[[0 for _ in range(V + 1)] for _ in range(W + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        for w in range(W + 1):
            for v in range(V + 1):


                if i == 0 or w == 0 or v == 0:
                    dp[i][w][v] = 0

                elif wt[i-1] <= w and vol[i-1] <= v:
                    dp[i][w][v] = max(
                        val[i-1] + dp[i-1][w - wt[i-1]][v - vol[i-1]], # Include
                        dp[i-1][w][v]                                  # Exclude
                    )

                else:
                    dp[i][w][v] = dp[i-1][w][v]

    return dp[n][W][V]
```

```python
#Memoization

calc = [[[-1 for _ in range(V + 1)] for _ in range(W + 1)] for _ in range(n + 1)]

def memoization(w, v, wt, vol, val, n):
    if n == 0 or w == 0 or v == 0:
        return 0

    # Check 3D Memo Table
    if calc[n][w][v] != -1:
        return calc[n][w][v]

    if wt[n-1] > w or vol[n-1] > v:
        calc[n][w][v] = memoization(w, v, wt, vol, val, n-1)
    else:
        calc[n][w][v] = max(
            val[n-1] + memoization(w - wt[n-1], v - vol[n-1], wt, vol, val, n-1),
            memoization(w, v, wt, vol, val, n-1)
        )

    return calc[n][w][v]
```

```python
print("Recursion Result:", recursion(W, V, wt, vol, val, n))
print("Bottom-Up DP Result:", dynamicprogramming(W, V, wt, vol, val, n))
print("Memoization Result:", memoization(W, V, wt, vol, val, n))
```

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
def fib(n):

    if n <= 1:
        return n

    dp = [0] * (n + 1)

    dp[0] = 0
    dp[1] = 1

    for i in range(2, n + 1):

        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

n = 4
print(f"Fibonacci number at position {n} is: {fib(n)}")
```

* Use recursion and dynamic programming to solve the problem

```python
#Exam room allocation using recursion

def exam_room (seats, sections_size, n):
  if seats == 0 or n == 0:
    return 0

  current_section = sections_size[n-1]

  if current_section > seats:
    return exam_room(seats, sections_size, n-1)

  else:
    include = current_section + exam_room(seats - current_section, sections_size, n-1)
    exclude = exam_room(seats, sections_size, n-1)
    return max(include, exclude)

seats = 70
sections_size = [10, 15, 20, 25, 30, 35]
n = len(sections_size)

used_seats = exam_room(seats, sections_size, n)
print(f"Max students seated: {used_seats} / {seats}")
```

```python
#Exam room allocation using dynamic programming

def examination(seat_available, section_size):
    n = len(section_size)
    dp = [[0 for _ in range(seat_available + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        for w in range(seat_available + 1):
            if i == 0 or w == 0:
                dp[i][w] = 0

            current_section_size = section_size[i-1]

            if current_section_size <= w:
                dp[i][w] = max(
                    current_section_size + dp[i-1][w - current_section_size],
                    dp[i-1][w]
                )
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][seat_available]

seat_available = 70
section_size = [10, 15, 20, 25, 30, 35]

used_seats = examination(seat_available, section_size)
print(f"Max students seated: {used_seats} / {seat_available}")
```
